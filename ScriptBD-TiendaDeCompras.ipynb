# [dbo].[UsuarioBorrarProducto]

```sql
/****** Object:  StoredProcedure [dbo].[UsuarioBorrarProducto]    Script Date: 15/05/2022 0:18:27 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[UsuarioBorrarProducto](
@idProducto int

)AS
BEGIN



 IF(SELECT count(*) FROM Producto where id= @idProducto)=0
 BEGIN 
		RAISERROR('Producto no encontrado',16,1);
		RETURN -1
 END

 IF(SELECT count(*) FROM DetalleVenta WHERE @idProducto=@idProducto)>0
  BEGIN 
		RAISERROR('Un Producto con Detalle de Venta no puede borrarse',16,1);
		RETURN -1
 END

-- Se eliminan las registros correspiondientes a las tablas Producto-DetalleVenta relacionadas
DELETE FROM dbo.DetalleVenta WHERE @idProducto=@idProducto

DELETE FROM Producto WHERE Id=@idProducto

END
GO
```

# [dbo].[UsuarioBuscarProductosTopMasVendidos]

```sql
/****** Object:  StoredProcedure [dbo].[UsuarioBuscarProductosTopMasVendidos]    Script Date: 15/05/2022 0:18:27 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[UsuarioBuscarProductosTopMasVendidos]

AS
BEGIN

SELECT TOP 3 P.NombreProducto 'PRODUCTO', MAX((DV.CantidadDetalleVenta)) 'PRODUCTOS MAS VENDIDOS', (SUM(V.TotalVenta)) VENTA
FROM Venta V INNER JOIN DetalleVenta DV ON V.Id = DV.Id 
             INNER JOIN Producto P ON p.Id = DV.Id  
GROUP  BY  P.NombreProducto, DV.CantidadDetalleVenta, V.TotalVenta
ORDER  BY V.TotalVenta DESC 



END
GO
```

# [dbo].[UsuarioBuscarProductosTopMenosVendidos]

```sql
/****** Object:  StoredProcedure [dbo].[UsuarioBuscarProductosTopMenosVendidos]    Script Date: 15/05/2022 0:18:27 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[UsuarioBuscarProductosTopMenosVendidos]

AS
BEGIN

SELECT TOP 3 P.NombreProducto, MIN(DV.CantidadDetalleVenta) 'MENOS VENDIDOS', SUM(TotalVenta) VENTA
FROM Venta V INNER JOIN DetalleVenta DV ON V.Id = DV.Id 
             INNER JOIN Producto P ON p.Id = DV.Id
GROUP  BY P.NombreProducto, CantidadDetalleVenta, V.TotalVenta
ORDER  BY V.TotalVenta ASC;



END
GO
```

# [dbo].[UsuarioModificaProducto]

```sql
/****** Object:  StoredProcedure [dbo].[UsuarioModificaProducto]    Script Date: 15/05/2022 0:18:27 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[UsuarioModificaProducto](
@Nombre varchar(500),
@Descripcion varchar(500),
@Precio decimal(11,2),
@Imagen varchar(500),
@Cantidad integer,
@FechaAlta datetime=null,
@IdProducto int =null

)AS
BEGIN

IF(@IdProducto IS NULL)
	BEGIN

		INSERT INTO dbo.Producto(NombreProducto,DescripcionProducto,PrecioProducto,ImagenProducto,CantidadProducto)
		VALUES(@Nombre,@Descripcion,@Precio,@Imagen,@Cantidad)

		SELECT @@IDENTITY as idProducto
	END
ELSE
BEGIN
	UPDATE dbo.Producto SET NombreProducto = @Nombre ,DescripcionProducto = @Descripcion,ImagenProducto = @Imagen,CantidadProducto = @Cantidad
	WHERE id=@IdProducto

	SELECT @IdProducto as idProducto
END


END
GO
```
